Histograms in Plotly using graph_objects class 

```python
import plotly.graph_objects as go 
import plotly.express as px 

df = px.data.iris() 
fig = go.Figure() 
fig.add_trace(go.Histogram(x=df['sepal_width'])) 
fig.add_trace(go.Histogram(x=df['species_id'])) 
fig.update_layout(barmode='stack') 
fig.show() 
```

```python
language=['Python','R language','Tableau',"Power BI",'SQL'] 
marks=[80, 75, 42, 78, 60,68] 
fig=go.Figure([go.Bar(x=language,y=marks,name='aws'),go.Bar(x=language,y=[60,67,45,34,78,90],name='gcs')]) 
 
fig.show() 
```

To create a stacked bar chart,

```python
import plotly.graph_objects as go 
language=['Python','R language','Tableau',"Power BI",'SQL'] 
marks=[80, 75, 42, 78, 60,68] 
fig=go.Figure([go.Bar(x=language,y=marks,name='aws'),go.Bar(x=language,y=[60,67,45,34,78,90],name='gcs')]) 
fig.update_layout(barmode='stack') 
fig.show() 
```

To sort the values within a bar chart,categoryorder ascending and similar 
for descending order to 

```python
import plotly.graph_objects as go 
language=['Python','R language','Tableau',"Power BI",'SQL'] 
marks=[80, 75, 42, 78, 60,68] 
fig=go.Figure([go.Bar(x=language,y=marks,name='aws'),go.Bar(x=language,y=[60,67,45,34,78,90],name='gcs')]) 
fig.update_layout(barmode='stack',xaxis={'categoryorder': "total ascending"}) 
fig.show()
```

Pie chart 

```python
import plotly.graph_objects as go 
language=['Python','R language','Tableau',"Power BI",'SQL'] 
marks=[80, 75, 42, 78, 60,68] 
 
trace=go.Pie(labels=language,values=marks) 
data=[trace] 
 
fig=go.Figure(data=data) 
fig.show() 
```

update traces and line width and color 

```python
import plotly.graph_objects as go 
language=['Python','R language','Tableau',"Power BI",'SQL'] 
marks=[80, 75, 42, 78, 60,68] 
colors = ['darkred', 'magenta', 'cyan', 'darkorange', 'darkblue', 'gold'] 
 
traces=go.Pie(labels=language,values=marks) 
 
data=[traces] 
fig=go.Figure(data=data) 
 
fig.update_traces(marker=dict(colors=colors,line=dict(color='black',width=2))) 
 
fig.show() 
```

Showing the Text Inside Pie Sectors 

```python
import plotly.graph_objects as go 
language=['Python','R language','Tableau',"Power BI",'SQL'] 
marks=[80, 75, 42, 78, 60,68] 
colors = ['darkred', 'magenta', 'cyan', 'darkorange', 'darkblue', 'gold'] 
 
traces=go.Pie(labels=language,values=marks,textinfo='label+percent') 
 
data=[traces] 
fig=go.Figure(data=data) 
 
fig.update_traces(marker=dict(colors=colors,line=dict(color='black',width=2))) 
 
fig.show() 
```

```python
import plotly.graph_objects as go 
language=['Python','R language','Tableau',"Power BI",'SQL'] 
marks=[80, 75, 42, 78, 60,68] 
colors = ['darkred', 'magenta', 'cyan', 'darkorange', 'darkblue', 'gold'] 
 
traces=go.Pie(labels=language,values=marks,textinfo='label+percent',hole=.5,insidetextorientation='radial') 
 
data=[traces] 
fig=go.Figure(data=data) 
 
fig.update_traces(marker=dict(colors=colors,line=dict(color='black',width=2))) 
 
fig.show() 
```

```python
import plotly.graph_objects as go 
language=['Python','R language','Tableau',"Power BI",'SQL'] 
marks=[80, 75, 42, 78, 60,68] 
colors = ['darkred', 'magenta', 'cyan', 'darkorange', 'darkblue', 'gold'] 
traces=go.Pie(labels=language,values=marks,textinfo='label+percent',pull=[0.0,0.1,0.4]) 
data=[traces] 
fig=go.Figure(data=data) 
fig.update_traces(marker=dict(colors=colors,line=dict(color='black',width=2))) 
fig.show() 
```
